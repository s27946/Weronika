Zadanie 1 (5 pkt)


Wgraj zawartość pliku orders.csv do ramki danych.
Przyjrzyj się danym, wykorzystując poznane metody (describe, info, head) zanalizuj dane.

Następnie:



1. Zamień order_date na format daty. Upewnij się, że w rezultacie data będzie w formacie: rok-miesiąc-dzień;
2. Sprawdź ile i jakie unikalne wartości znajdują się w kolumnie T-shirt category;
3. Wyczyść kolumnę tshirt_category normalizując nazwy produktów. Upewnij się, że dla każdej kategorii funkcjonuje jedna nazwa, np. zamiast white i wh jest tylko jedna wersja white. Skorzystaj z klasy string i metody apply;
4. Rozdziel kolumnę tshirt category na 3 kolumny: tshirt_type, która będzie mówiła o tym czy jest to koszulka, bluza czy coś innego, tshirt_size, która będzie informowała o rozmiarze koszulki, i tshirt_colour, która będzie informowała o kolorze. Skorzystaj z metody apply.
5. Znajdź zamówienia złożone między 31 grudnia 2014 a 2 sierpnia 2016.


```python
import pandas as pd
orders = pd.read_csv('orders.csv', encoding='utf-8', sep=',')

#zad1
orders['order_date'] = orders['order_date'].apply(lambda x: pd.to_datetime(x,errors = 'coerce', format = '%Y-%m-%d'))
orders.head(3)
orders.describe()
orders.info()

#zad2 
orders['tshirt_category'].unique()
orders['tshirt_category'].nunique()

#zad3
mask_hoodie = (orders['tshirt_category'] == 'Hoodie')
mask_tennis_tshirt = (orders['tshirt_category'] == 'Tennis Shirt')
#dopisanie brakujących danych dla Hoodie
orders.loc[mask_hoodie, 'tshirt_category'] = 'Black ' + orders['tshirt_category'].astype(str) + ' XL'
#Tennis Shirt jest teraz odczytany jako jeden string, tj. jeden typ produktu - Tennis_Shirt
orders.loc[mask_tennis_tshirt, 'tshirt_category'] = orders['tshirt_category'].str.replace(' ','_')
#dopisanie brakujących danych dla Tennis Shirt
orders.loc[mask_tennis_tshirt, 'tshirt_category'] = 'Yellow ' +  orders['tshirt_category'].astype(str) + ' M'
#czyszczenie błędnych danych
values_to_update = {"tshirt_category": 
                  {'Bl Tshirt F':'Black T-Shirt F', 'Bl Tshirt M':'Black T-Shirt M',
                   'Wh Tshirt F': 'White T-Shirt F', 'Wh Tshirt M': 'White T-Shirt M' }}
orders.replace(values_to_update, inplace=True)
set(orders['tshirt_category'])

#zad4

orders['tshirt_colour']=orders['tshirt_category'].str.split().apply(lambda x: x[0])
orders['tshirt_type']=orders['tshirt_category'].str.split().apply(lambda x: x[1]) 
orders['tshirt_size']=orders['tshirt_category'].str.split().apply(lambda x: x[-1])
orders.head(10)

#zad5
mask_date = (orders['order_date'] < '2016-08-02') & (orders['order_date'] > '2014-12-31')
orders[mask_date].sort_values(by=["order_date"])
orders.head(10)
```
